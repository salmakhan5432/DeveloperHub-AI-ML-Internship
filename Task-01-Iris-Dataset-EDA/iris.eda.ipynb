# Task 01: Iris Dataset - Exploratory Data Analysis


```python
# Import pandas for data handling
import pandas as pd
# Import seaborn and matplotlib for  
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset form seaborn library
iris = sns.load_dataset("iris")
```

```python
# Check shape(rows, columns)
print("Shape of dataset: ",iris.shape)
```

```python
# Check column names
print("Column names:", iris.columns)
```

```python
# Show first five rows to check data 
iris.head()
```

```python
# Detailed info: data types, non-null counts
iris.info()
```

```python
# Summary statistics: mean, min, max, std
iris.describe()
```

```python
# Scatter plot to show relationship between sepal length and sepal width
# Hue is species to color code each type of Iris
plt.figure(figsize= (6,4))
sns.scatterplot(
    data = iris,
    x = "sepal_length",
    y = "sepal_width",
    hue = "species"
)
plt.title("Sepal Length vs sepal width")
plt.show()
```

```python
# Histograms to see distribution of each numerical feature
iris.hist(figsize=(8,6))
plt.suptitle("Feature Distribution in Iris Dataset")
plt.show()
```

```python
# Box plots to detect outliers in each feature
plt.figure(figsize = (8,4))
sns.boxplot(data=iris)
plt.title("Box plot of Iris Features")
plt.show()
```

## Insights
- The Iris dataset contains 150 samples with 4 numerical features.
- Petal length and petal width show clear separation between species.
- No major missing values are present in the dataset.
- Some features show mild outliers, visible in box plots.
